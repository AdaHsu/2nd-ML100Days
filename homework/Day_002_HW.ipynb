#### 資料的 row 數以及 column 數

#### 截取部分資料

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
# 資料過濾 - 找大於 70 歲的資料
over70_index = df[df.AGE > 70]
print(type(over70_index))
print('大於 70 歲的人數有：{0} 人'.format(len(over70_index)))
over70_index[['ID','LIMIT_BAL','SEX','EDUCATION','MARRIAGE','AGE','default.payment.next.month']]
```

```python
# 排除 最近 2 個月都沒有使用信用卡資料
no_credit=df[(df.PAY_0 == -2) & (df.PAY_2 == -2)]
no_credit.shape
```

```python
# 剩除最近 2 個月使用信用卡的資料
using_credit=df[(df.PAY_0 != -2) | (df.PAY_2 != -2)]
using_credit.shape
```

```python
# 將年齡轉成類別型資料, 自 20 歲開始每 5 年為一個單位, 上限 70+
sections = [0,19,24,29,34,39,44,49,54,59,64,69,np.inf]
# 類別 (一般會設定為數值)
category_labels = ['<20', '20~24', '25~29', '30~34','35~39','40~44', '45~49', '50~54', '55~59', '60~64', '65~69', '70+']
# using_credit.assign(AGE_CATEGORY = pd.cut(using_credit['AGE'], sections, labels=category_labels))
using_credit.insert(6, 'AGE_CATEGORY', pd.cut(using_credit['AGE'], sections, labels=category_labels), True)
using_credit
```

```python
using_credit[['ID','AGE','AGE_CATEGORY']]
```
