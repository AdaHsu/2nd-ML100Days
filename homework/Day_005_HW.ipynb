```python
app_train.describe()
```

```python
print( app_train['AMT_CREDIT'].max())
print( app_train['AMT_CREDIT'].min())
app_train['AMT_CREDIT'].describe()
```

```python
# 貧富差距甚大
print( app_train['AMT_INCOME_TOTAL'].max())
print( app_train['AMT_INCOME_TOTAL'].min())
print( app_train['AMT_INCOME_TOTAL'].quantile(0.25))
app_train['AMT_INCOME_TOTAL'].describe()
```

```python
# 濾出收入小於第一四分位數的記錄
less_q1 = app_train.loc[app_train.AMT_INCOME_TOTAL < app_train['AMT_INCOME_TOTAL'].quantile(0.25), ['AMT_CREDIT', 'AMT_INCOME_TOTAL', 'AMT_GOODS_PRICE']]
less_q1
```

```python
less_q1.hist(layout=(3,1))
less_q1.plot.density()
less_q1.plot.box()
```

```python
plt.scatter(less_q1.AMT_INCOME_TOTAL, less_q1.AMT_CREDIT)
plt.show()
```

```python
# 貸款金額大扺和購買的商品價格相近
plt.scatter(less_q1.AMT_GOODS_PRICE, less_q1.AMT_CREDIT)
plt.show()
```

```python
# 收入愈多，購買的商品愈高檔，但借的錢也多
from pandas.plotting import scatter_matrix
scatter_matrix(less_q1, alpha=0.2, figsize=(20, 20), diagonal='kde') 
```

```python
# 濾出收入 >= 中位數
over_q2 = app_train.loc[app_train.AMT_INCOME_TOTAL >= app_train['AMT_INCOME_TOTAL'].quantile(0.5), ['AMT_CREDIT', 'AMT_INCOME_TOTAL', 'AMT_GOODS_PRICE']]
over_q2
```

```python
# 呃...
scatter_matrix(over_q2, alpha=0.2, figsize=(20, 20), diagonal='kde') 
```
